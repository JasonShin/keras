What is it?

### 1. Features from text

- Given a block of text
- Come up with word:index pairs

(the, 1) (be, 2) (to, 3) (of, 4)

Problem: Numbers will become very large

### Solution: Embedding Layer

- Takes the huge sparse vector and full connects it into much smaller nodes
- Makes the words dense
- Encapsulates semantic meaning
- Produces semantic cloud of vectors
- We can perform mathmatical operation with words
        King - Man + Woman = Queen

```python
# why pad?  views have different lengths
from keras.preprocessing.sequence import pad_sequences
from keras.layers import LSTM
```

```python
maxlen = 100
X_train_pad = pad_sequences(X_train, maxlen=maxlen)
X_test_pad = pad_sequences(X_test, maxlen=maxlen)
```

```python
X_train_pad.shape
```

```python
X_train_pad[0]
```

```python
X_train[0]
```

```python
# checking max features
max_features = max([max(x) for x in X_train_pad] + [max(x) for x in X_test_pad]) + 1
max_features
```

```python
model = Sequential()
model.add(Embedding(max_features, 128))
model.add(LSTM(64, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train_pad, y_train,
         batch_size=32,
         epochs=2,
         validation_split=0.3)
```

```python
score, acc = model.evaluate(X_test_pad, y_test)
print('Test score:', score)
print('Test accuracy:', acc)
```
